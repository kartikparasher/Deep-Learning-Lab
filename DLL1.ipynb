```python
import skimage as ski

image = ski.io.imread('Screenshot 2024-08-08 at 8.43.37 AM.jpg')  # Replace with your image path

# Check if the image has an alpha channel
if image.shape[-1] == 4:
    # Remove the alpha channel
    image = image[:, :, :3]

# Now you can use functions that expect RGB images
edges = ski.filters.sobel(ski.color.rgb2gray(image))
ski.io.imshow(edges)
ski.io.show()
```
